```python


# Criar cliente local persistente
client = chromadb.PersistentClient(path="./clip_chroma")

# Criar coleção (sem embedding_function)
colecao = client.get_or_create_collection(name="meus_docs_clip")

# ---- Exemplo de embeddings já gerados pelo CLIP ----
# Digamos que você já tem os embeddings (cada linha = 1 documento)
docs = [
    "Um gato sentado no sofá",
    "Um cachorro correndo no parque",
    "Um notebook sobre a mesa"
]

# Suponha que já rodou CLIP e tem os vetores
# (aqui só crio fake embeddings para ilustrar)
embeddings_clip = np.random.rand(3, 512).tolist()  # CLIP ViT-B/32 -> 512 dim

# ---- Adicionar ao Chroma ----
colecao.add(
    documents=docs,
    embeddings=embeddings_clip,   # passa direto
    ids=[f"doc{i}" for i in range(len(docs))]
)

print("Documentos salvos no ChromaDB!")
```

```python
# Reabrir o banco
client = chromadb.PersistentClient(path="./clip_chroma")

# Carregar a coleção
colecao = client.get_collection(name="meus_docs_clip")

# Pegar tudo de uma vez
todos = colecao.get(
    include=[ "embeddings"]  # você escolhe o que trazer
)


print("Embeddings:", todos["embeddings"])  # quantos vetores vieram
print("Dimensão de cada vetor:", len(todos["embeddings"][0]))
```
